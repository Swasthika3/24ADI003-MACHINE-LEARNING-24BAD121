```python
print("SWASTHIKA M 24BAD121")
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import networkx as nx
data = pd.read_csv("Groceries_dataset.csv")
print("Dataset Preview:\n", data.head())

transactions = data.groupby('Member_number')['itemDescription'].apply(list).values.tolist()

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)

print("\nOne-Hot Encoded Data:\n", df.head())
# Apriori Algorithm
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)
print("\nFrequent Itemsets:\n", frequent_itemsets.head())
# Association Rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
print("\nAssociation Rules:\n", rules.head())
strong_rules = rules[(rules['lift'] > 1) & (rules['confidence'] > 0.3)]
print("\nStrong Rules:\n", strong_rules.head())
# Visualization 1: Bar Chart

top_items = frequent_itemsets.sort_values(by='support', ascending=False).head(10)

plt.figure()
sns.barplot(x=top_items['support'], y=top_items['itemsets'].astype(str))
plt.title("Top Frequent Itemsets")
plt.xlabel("Support")
plt.ylabel("Itemsets")
plt.show()
# Visualization 2: Support vs Confidence

plt.figure()
sns.scatterplot(x='support', y='confidence', size='lift', data=rules)
plt.title("Support vs Confidence")
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.show()
# Visualization 3: Network Graph

G = nx.DiGraph()

for _, row in strong_rules.head(15).iterrows():
    for antecedent in row['antecedents']:
        for consequent in row['consequents']:
            G.add_edge(antecedent, consequent, weight=row['confidence'])

plt.figure(figsize=(8,6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, font_size=8)
plt.title("Association Rules Network")
plt.show()
# Interpretation
print("\nInterpretation:")
print("• Frequently bought items have high support")
print("• Rules with higher confidence are more reliable")
print("• Lift > 1 indicates strong association between items")
```

```python
print("SWASTHIKA M 24BAD121")
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

print("\nDataset Preview:\n", df.head())
print("\nMissing Values:\n", df.isnull().sum())

# Features & target
X = df.drop('target', axis=1)
y = df['target']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Apply PCA

pca = PCA()
X_pca = pca.fit_transform(X_scaled)
# Explained Variance

explained_variance = pca.explained_variance_ratio_
print("\nExplained Variance Ratio:\n", explained_variance)

cumulative_variance = explained_variance.cumsum()
print("\nCumulative Variance:\n", cumulative_variance)
# Reduce to 2 Components
pca2 = PCA(n_components=2)
X_reduced = pca2.fit_transform(X_scaled)

reduced_df = pd.DataFrame(X_reduced, columns=['PC1', 'PC2'])
reduced_df['target'] = y

print("\nReduced Data:\n", reduced_df.head())
# Visualization 1: Scree Plot

plt.figure()
plt.plot(range(1, len(explained_variance)+1), explained_variance, marker='o')
plt.title("Scree Plot")
plt.xlabel("Principal Components")
plt.ylabel("Variance")
plt.show()
# Visualization 2: Cumulative Variance

plt.figure()
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.title("Cumulative Variance")
plt.xlabel("Components")
plt.ylabel("Cumulative Variance")
plt.show()
# Visualization 3: 2D Scatter Plot

plt.figure()
sns.scatterplot(x='PC1', y='PC2', hue='target', data=reduced_df)

plt.title("PCA 2D Scatter Plot")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

print("\nInterpretation:")
print("• PCA reduces dimensions while keeping maximum variance")
print("• First few components capture most information")
print("• Helps in visualization and faster computation")
```
